```python
import pandas as pd
import matplotlib.pyplot as plt


class RetailSalesAnalyzer:
    def __init__(self, data_file):
        self.data_file = pd.read_csv(data_file)
        self.data_file['Date'] = pd.to_datetime(self.data_file['Date'])

    def Data_cleaning(self):
        self.data_file.dropna(inplace=True)

    def Calculate_total_sales(self):
        return self.data_file.groupby('Product')['Sales'].sum()

    def Best_selling_product(self):
        return self.Calculate_total_sales().sort_values(ascending=False).head(1)

    def Average_daily_sales(self):
        return self.data_file['Sales'].mean()

    def plot_sales_Trends(self):
        self.data_file.groupby('Date')['Sales'].sum().plot(kind='line', title='Sales Trends Over Time')
        plt.xlabel('Date')
        plt.ylabel('Total Sales')
        plt.show()

    def Plot_Sales_per_product(self):
        self.Calculate_total_sales().plot(kind='bar', title='Total Sales per Product')
        plt.xlabel('Product')
        plt.ylabel('Total Sales')
        plt.show()

```

```python
analyzer = RetailSalesAnalyzer('/content/sample_data/retail_sales.csv')
print('Total Sales per Product:\n', analyzer.Calculate_total_sales())
print('Best selling Products:\n', analyzer.Best_selling_product())
print('Average Daily Sales:', analyzer.Average_daily_sales())
analyzer.Plot_Sales_per_product()
analyzer.plot_sales_Trends()
```
